# **Data Cleaning**

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#  Here some cols. values are null
d = df["Displacement"].mean()
df["Displacement"].fillna(d, inplace = True)

# transmission

df[df["Model"] == "Kona Electric"]

# we drop 155th index row

df.drop(df.index[155], inplace=True)


```

```python
df.isnull().sum()
```

```python
#  rename cols.

df.rename(columns={"Engine Type":"Engine_Type",
                   "Front Brake":"Front_Brake",
                   "Rear Brake":"Rear_Brake",
                   "Turning Radius":"Turning_Radius",
                   "Fuel Tank Capacity":"Fuel_Tank_Capacity",
                   "Boot Space":"Boot_Space",
                   "Fuel Efficiency":"Fuel_Efficiency",
                   "Emission Type":"Emission_Type",
                   "Tyre Size":"Tyre_Size",
                   "NCAP Rating":"NCAP_Rating"},inplace=True)
```

```python
R = df["Turning_Radius"].mean()
df["Turning_Radius"].fillna(R , inplace =True)


B = df["Boot_Space"].mean()
df["Boot_Space"].fillna(B, inplace =True)

E = df["Fuel_Efficiency"].mean()
df["Fuel_Efficiency"].fillna(E, inplace =True)
```
