```python
df.rename(columns={"Engine Type":"Engine_Type",
                   "Front Brake":"Front_Brake",
                   "Rear Brake":"Rear_Brake",
                   "Turning Radius":"Turning_Radius",
                   "Fuel Tank Capacity":"Fuel_Tank_Capacity",
                   "Boot Space":"Boot_Space",
                   "Fuel Efficiency":"Fuel_Efficiency",
                   "Emission Type":"Emission_Type",
                   "Tyre Size":"Tyre_Size",
                   "NCAP Rating":"NCAP_Rating"},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
x=df.No_of_Cylinders.mean()
df.No_of_Cylinders.fillna(x,inplace=True)
```

```python
a=df.Displacement.mean()
df.Displacement.fillna(a,inplace=True)
```

```python
b=df.Turning_Radius.mean()
df.Turning_Radius.fillna(b,inplace=True)
```

```python
c=df.Boot_Space.mean()
df.Boot_Space.fillna(c,inplace=True)
```

```python
d=df.Fuel_Efficiency.mean()
df.Fuel_Efficiency.fillna(d,inplace=True)
```

```python
df.Emission_Type.dropna(inplace=True)
```

## Data Visulization / EDA

```python
sns.pairplot(df)
plt.show()
```

```python
cormat=df.corr()
top_corr_features = cormat.index
plt.subplots(figsize=(9, 9))
sns.heatmap(df[top_corr_features].corr(), annot = True, cmap ="mako",fmt='.2')
plt.show()
```

```python
data1 = df['Model']
data2= df['Maker']
conc_data_col = pd.concat([data1,data2],axis =1) # axis = 1 : adds dataframes in column
conc_data_col
```

```python
a,ax=plt.subplots(figsize=(12,12))
sns.countplot(data=df,x="Model", saturation=.75, dodge=False,ax=ax,palette="terrain")
plt.xticks(rotation=90)
plt.xlabel("Car's Model",fontsize=20,color="blue")
plt.show()
```

```python
labels = df["Model"].value_counts().index
sizes = df["Model"].value_counts()
plt.figure(figsize = (10,10))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%')
plt.title('Model',color = 'purple',fontsize = 15)
plt.show()
```
